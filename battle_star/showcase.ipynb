```python
from battle_star import generate_puzzle, solve_puzzle, draw_puzzle_inline
```

Let us generate a puzzle

```python
grid_size, nstars = 10, 2

region_grid, star_grid = generate_puzzle(grid_size, nstars, annealer="leap")
draw_puzzle_inline(region_grid, star_grid)
```

Let us now solve the problem for this region, using annealing as well!

```python
solution_grid = solve_puzzle(grid_size, nstars, region_grid, annealer="neal", num_reads=200)
draw_puzzle_inline(region_grid, solution_grid)
```
